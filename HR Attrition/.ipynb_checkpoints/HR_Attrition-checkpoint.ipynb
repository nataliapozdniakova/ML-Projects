## Univariate analysis


```python
import matplotlib.pyplot as plt
import seaborn as sns

cat_vars = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']

n = len(cat_vars)
cols = 3
rows = (n + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(5*cols, 5*rows))
axes = axes.flatten()

for i, col in enumerate(cat_vars):
    order = df[col].value_counts().index
    ax = sns.countplot(x=col, data=df, ax=axes[i], color='seagreen', order=order)

    total = len(df)
    for p in ax.patches:
        percentage = f'{100 * p.get_height()/total:.1f}%'
        ax.annotate(percentage, (p.get_x() + p.get_width()/2., p.get_height()),
                    ha='center', va='bottom', fontsize=10)
    
    ax.set_title(f'Distribution of {col}', fontsize=12)
    ax.tick_params(axis='x', rotation=45)

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
num_vars = [
    'Age',
    'DailyRate',
    'DistanceFromHome',
    'HourlyRate',
    'MonthlyIncome',
    'MonthlyRate',
    'NumCompaniesWorked',
    'PercentSalaryHike',
    'TotalWorkingYears',
    'TrainingTimesLastYear',
    'YearsAtCompany',
    'YearsInCurrentRole',
    'YearsSinceLastPromotion',
    'YearsWithCurrManager'
]

print("Continuous numeric variables for analysis:", num_vars)
```

```python
cols = 3
rows = (len(num_vars) + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(6*cols, 5*rows))
axes = axes.flatten()

for i, col in enumerate(num_vars):
    sns.histplot(df[col], kde=True, ax=axes[i], color='skyblue', bins=20)
    axes[i].set_title(f'Distribution of {col}', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
ordinal_vars = [
    'Education',            
    'EnvironmentSatisfaction',  
    'JobInvolvement',       
    'JobLevel',             
    'JobSatisfaction',      
    'PerformanceRating',   
    'RelationshipSatisfaction',
    'StockOptionLevel',    
    'WorkLifeBalance'      
]


import matplotlib.pyplot as plt
import seaborn as sns

n = len(ordinal_vars)
cols = 3
rows = (n + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(5*cols, 5*rows))
axes = axes.flatten()

for i, col in enumerate(ordinal_vars):
    order = sorted(df[col].unique())  
    ax = sns.countplot(x=col, data=df, ax=axes[i], color='lightcoral', order=order)
    
    total = len(df)
    for p in ax.patches:
        percentage = f'{100 * p.get_height()/total:.1f}%'
        ax.annotate(percentage, (p.get_x() + p.get_width()/2., p.get_height()),
                    ha='center', va='bottom', fontsize=10)
    
    ax.set_title(f'Distribution of {col}', fontsize=12)

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()



```

## Bivariate analysis

Numerical variables

```python
def numcol_viz(col_name):
    attrition_df = df[df['Attrition'] == 'Yes']
    no_attrition_df = df[df['Attrition'] == 'No']

    colors = {'Yes': 'salmon', 'No': 'skyblue'}

    f, ax = plt.subplots(1, 2, figsize=(16, 5))

    sns.kdeplot(attrition_df[col_name], fill=True, label='Left', color=colors['Yes'], ax=ax[0])
    sns.kdeplot(no_attrition_df[col_name], fill=True, label='Stayed', color=colors['No'], ax=ax[0])
    ax[0].set_title(f'Distribution of {col_name}', fontsize=14)
    ax[0].legend()

    sns.boxplot(
        x='Attrition',
        y=col_name,
        color='skyblue',
        data=df,
        ax=ax[1],
        width=0.4 
    )
    ax[1].set_title(f'{col_name} by Attrition', fontsize=14)

    plt.tight_layout()
    plt.show()

for col in num_vars:
    numcol_viz(col)


```


Categorical variables

```python
df['Attrition_label'] = df['Attrition'].map({'No': 'Stayed', 'Yes': 'Left'})

def catcol_viz(col_name):
    plt.figure(figsize=(6,4))
    
    colors = {'Stayed': 'skyblue', 'Left': 'salmon'}
    
    ax = sns.countplot(
        x=col_name,
        hue='Attrition_label',
        data=df,
        palette=colors
    )
    
    plt.title(f'{col_name} Count by Attrition', fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

for col in cat_vars:
   catcol_viz(col)
```

```python
df['Attrition_label'] = df['Attrition'].map({'No': 'Stayed', 'Yes': 'Left'})

def ordinal_viz(col_name):
    
    plt.figure(figsize=(6,4))
    
    order = sorted(df[col_name].unique())
    colors = {'Stayed': 'skyblue', 'Left': 'salmon'}
    
    ax = sns.countplot(
        x=col_name,
        hue='Attrition_label',
        data=df,
        order=order,
        palette=colors
    )
    
    plt.title(f'{col_name} Count by Attrition', fontsize=14)
    plt.xlabel(col_name)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

for col in ordinal_vars:
    ordinal_viz(col)

```

```python
#I use Spearman corr matrix because there are ordinal values in the dataset

df['Attrition_num'] = df['Attrition'].map({'No': 0, 'Yes': 1})

corr_vars = num_vars + ordinal_vars + ['Attrition_num']

corr = df[corr_vars].corr(method='spearman')


plt.figure(figsize=(14,12))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title("Spearman Correlation Matrix", fontsize=16)
plt.show()

```
